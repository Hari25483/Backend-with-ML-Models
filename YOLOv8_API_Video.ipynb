```python
from flask import Flask,request, jsonify
from flask_ngrok import run_with_ngrok
from flask_cors import CORS, cross_origin
from ultralytics import YOLO
import urllib
from moviepy.editor import VideoFileClip
model = YOLO("/content/drive/MyDrive/predict_model/best.pt")  # load a pretrained model (recommended for training)

def compress(path):
    # specify the input and output file paths
    input_file_path = path
    output_file_path = "/content/a.mp4"

    # load the video clip
    video_clip = VideoFileClip(input_file_path)

    # reduce the video resolution and frame rate
    video_clip_resized = video_clip.resize((480, 360))

    # write the compressed video to file
    video_clip_resized.write_videofile(output_file_path, codec='libx264')

app = Flask(__name__)
CORS(app)
run_with_ngrok(app)


@app.route('/')
def index():
    return jsonify("Welcome to the home page")

@app.route('/classify', methods=['POST'])
def upload_audio():
    if request.method == 'POST':
        f = request.files['video']
        name=f.filename
        f.save(f.filename)
        print(f.filename)
        path="/content/"+str(f.filename)
        print(path)
        results = model.predict(source=path, save=True, save_txt=True)
        path1="/content/runs/detect/predict/"+str(name)
        print(path1)
        compress(path1)
        #upload a file
        file="/content/a.mp4"
        cloudfilename="a"
        storage.child(cloudfilename).put(file)
        #get url of the file we just uploaded
        print(storage.child(cloudfilename).get_url(None))
        return jsonify(storage.child(cloudfilename).get_url(None))
    else:
        return 'No video in request'

app.run()
```
